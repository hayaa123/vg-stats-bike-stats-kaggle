````python
year_series = pd.to_datetime(bike_wheather["Date"]).dt.year
````

````python
month_series = pd.to_datetime(bike_wheather["Date"]).dt.month
````

````python
month_df = pd.DataFrame({"month":month_series,"year":year_series})
````

````python
group_by_mean = pd.concat([bike_wheather,month_df],axis=1)[["month","year","Precipitation_In"]].groupby(["year","month"]).mean()
group_by_mean
````

````python
group_by_median = pd.concat([bike_wheather,month_df],axis=1)[["month","year","Precipitation_In"]].groupby(["year","month"]).median()
group_by_median
````
